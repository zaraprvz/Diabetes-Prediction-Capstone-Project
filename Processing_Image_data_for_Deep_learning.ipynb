Resizing the image using Pillow library


```python
from PIL import Image
original = Image.open('/content/cat.jpg')
ime_resized = original.resize((200,200))
ime_resized.save('/content/cat_resized.jpg')
```

```python
img = mp.imread('/content/cat_resized.jpg')
img_plot = plt.imshow(img)
plt.show
```

```python
print(img.shape)
```

Converting RGB images to GrayScale image using OpenCV

```python
#import OpenCV library
import cv2
img = cv2.imread('/content/cat.jpg')
```

```python
type(img)
```

```python
grayscale_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

```
